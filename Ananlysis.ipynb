### Importing Libraries

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

### Loading the datasets

```python
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("Health_heart_experimental.csv")
df.head(10)
```

### Comparison of Sex Distribution

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

* As we can see from graph above, The number of males and females in the data set is nearly equal to each other.So, we can know the data set will likely to be true among both gender and not biased to one.

```python
# Bivariate analysis of SysBP by sex and indication
sns.barplot(x='sex', y='SysBP', hue='indication', data=df, palette='Set1')

# Add title and labels
plt.title("Average SysBP by Sex and Indication")
plt.xlabel("Sex")
plt.ylabel("Average SysBP")

# Show the plot
plt.show()
```

### Boxplot for Sex vs BMI

```python
sns.boxplot(x='sex', y='BMI', data=df)
plt.title('BMI by Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('BMI')
plt.show()
```

From the barplot we can observe the following things:<br>
* Both gender have identical distribution of BMI.
* There is no outlier in the graph.
* The BMI ranges from 15-40 and median is around 28.

### Plot of age vs bmi

```python
age = df['age']
bmi = df['BMI']
plt.plot(age, bmi, color='blue', marker='o', linestyle='solid')
plt.xlabel('Age')
plt.ylabel('BMI')
```

The data is too dense, possibly due to having a large number of overlapping data points. Solve it by using grouping.

```python

grouped_data = df.groupby('age')['BMI'].mean()

# Extract age and BMI for plotting
age = grouped_data.index
mean_bmi = grouped_data.values

plt.plot(age, mean_bmi, marker='o', linestyle='-', color='b')  # Line plot with markers
plt.title('BMI Across Age (Aggregated)')
plt.xlabel('Age')
plt.ylabel('Mean BMI')
```

From the dataset we can observe that, The age group of 29 has the lowest BMI of 28.55. The Data set projects the average person in the data set is overweught as BMI>28.

```python
sns.barplot(x='indication', y='age', data=df)
plt.title('Average Age by Indication')
plt.xlabel('Indication (0 = No, 1 = Yes)')
plt.ylabel('Average Age')
plt.show()
```
